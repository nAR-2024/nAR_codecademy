```python
#This is a basic ceasar cipher with a known shift used to decode a message

def ceasar_cipher_decode(text, shift):
    result=[]
    for char in text:
        if char.isupper():
            result += chr((ord(char) + shift - 65) % 26 + 65)
        elif char.islower():
            result += chr((ord(char) + shift - 97) % 26 + 97)
        else:
            result += char
    return result

plain_text = "jxu evviuj veh jxu iusedt cuiiqwu yi vekhjuud."
shift = 10
cypher_text = ceasar_cipher_decode(plain_text, shift)
print(plain_text)

print(cypher_text)
```

```python
#this is a basic ceasar cipher used to encode a meassage with a known shift.

def ceasar_cipher_encode(text, shift):
    result=[]
    for char in text:
        if char.isupper():
            result += chr((ord(char) - shift - 65) % 26 + 65)
        elif char.islower():
            result += chr((ord(char) - shift - 97) % 26 + 97)
        else:
            result += char
    return result

plain_text = "I wanted to make this where no one could read it"
shift = 10
cypher_text = ceasar_cipher_encode(plain_text, shift)
print(plain_text)

print(cypher_text)
```

```python
#this is a basic ceasar cipher used to encode a meassage with a known shift that 
#the user has the ability to enter the message and shift value.

def ceasar_cipher_userinput(text, shift):
    result=[]
    for char in text:
        if char.isupper():
            result += chr((ord(char) - shift - 65) % 26 + 65)
        elif char.islower():
            result += chr((ord(char) - shift - 97) % 26 + 97)
        else:
            result += char
    return result

plain_text = input("Enter the text to encrypt: ")
shift = int(input("Enter the shift amount: "))

# call the function
ciphertext = ceasar_cipher_userinput(plain_text, shift)
print(f"Encrypted message: {ciphertext}")
```

```python

# If you do not know the shift we can Brute force the decryption by trying all 25 possibilities 
# and see which one results in a readable message.

def brute_force_ceasar_cipher(ciphertext):
    for shift in range(1, 26): #trying all shifts from 1 to 25
        result = ""
        for char in ciphertext:
            if char.isupper():
                result += chr((ord(char) + shift - 65) % 26 + 65)
            elif char.islower():
                result += chr((ord(char) + shift - 97) % 26 + 97)
            else:
                result += char
        print(f"Shift {shift}: {result}")

ciphertext = input("Enter the text to decrypt: ") #encrypted message
brute_force_ceasar_cipher(ciphertext)
```

```python
#function to encrypt the plaintext using a vigenere cipher

def vigenere_encrypt(plaintext, keyword):
    keyword = keyword.upper()
    plaintext = plaintext.upper()
    cipher_text = []

    keyword_index = 0
    for letter in plaintext:
        if letter.isalpha():
            shift = ord(keyword[keyword_index]) - ord('A')
            encrypted_letter = chr(((ord(letter) - ord('A') + shift) % 26) + ord('A'))
            cipher_text.append(encrypted_letter)

            keyword_index = (keyword_index + 1) % len(keyword)
        else:
            cipher_text.append(letter)
    return''.join(cipher_text)

# Functin to decrypt the ciphertext using the Vigenere cipher

def vigenere_decrypt(ciphertext, keyword):
    keyword = keyword.upper()
    ciphertext = ciphertext.upper()
    decrypted_text = []

    keyword_index = 0
    for letter in ciphertext:
        if letter.isalpha():
            shift = ord(keyword[keyword_index]) - ord('A')
            decrypted_letter = chr(((ord(letter) - ord('A') - shift + 26) % 26) +ord('A'))
            decrypted_text.append(decrypted_letter)

            keyword_index = (keyword_index + 1) % len(keyword)
        else:
            decrypted_text.append(letter)
    return ''.join(decrypted_text)

# Main program to choose between encrypting and decrypting, and loops back around to allow a different selection

def main():
    while True:  #Allows for multiple sessions, instead of just one path

        while True:
            choice = input("Would you like to (E)ncrypt or (D)ecrypt your message? ").upper()
            if choice not in ['E', 'D']:
                break # breaks out until a valid choice is made, terrible to figure out
            else: 
                print("invalid choice.  Please enter 'E' for encrypt or 'D' for decrypt.")
        text = input("Enter the text: ")
        keyword = input("Enter the keyword: ")

        if choice == 'E':
            encrypted_text = vigenere_encrypt(text, keyword)
            print(f"Encrypted text: {encrypted_text}")
        elif choice == 'D':
            decrypted_text = vigenere_decrypt(text, keyword)
            print(f"Decrypted text: {decrypted_text}")

        # figure a way to ask the user for another session
        another_session = input("Do you want to start a new session? (Y?N): ").upper()
        if another_session != 'Y':
            print("Till the next time! ")
            break # exits the loop and ends the program
    
       

# Runs the Program
main ()


# get input from the user
plaintext = input("Enter the text: ")
keyword = input("Enter the keyword: ")

#encrypt the text
encrypted_text = vigenere_encrypt(plaintext, keyword)

#display it
print(f"Encrypted Text:  {encrypted_text}")
```
